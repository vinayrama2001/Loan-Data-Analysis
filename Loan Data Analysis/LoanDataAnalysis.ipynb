# **Logistic Regression**

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score

x =df.drop('not.fully.paid',axis=1)
y =df[['not.fully.paid']]
X_train, X_test, y_train, y_test = train_test_split(x,y, test_size=0.2,random_state=101)

```

```python
# FOR SCALING ------------------

# from sklearn.preprocessing import LabelEncoder,StandardScaler,RobustScaler
# sc= RobustScaler()
# X_train1=sc.fit_transform(X_train)
# X_test1=sc.transform(X_test)
```

```python
df_loan=  LogisticRegression()
df_loan.fit(X_train,y_train)
Y_pred = df_loan.predict(X_test)
X_pred = df_loan.predict(X_train)
print(classification_report(y_test,Y_pred))
print("------------------------\n")
print("Confusion Matrix : \n")
print(confusion_matrix(y_test,Y_pred))
```

```python
print("Accuracy for Training: ",accuracy_score(y_train,X_pred) )
print("Accuracy for Testing: ",accuracy_score(y_test,Y_pred))
```

```python
from sklearn.metrics import mean_squared_error
print("Mean squared error: %.2f"
      % mean_squared_error(y_test, Y_pred))
```

# **Random Forest Classifier**

```python
from sklearn.ensemble import RandomForestClassifier
from imblearn.over_sampling import SMOTE
from collections import Counter

rfc= RandomForestClassifier(random_state=0)
rfc.fit(X_train,y_train)

counter =Counter(y_train.value_counts())
print('Before',counter)
smt = SMOTE()
X_train_smt,y_train_smt =smt.fit_resample(X_train,y_train)
counter = Counter(y_train_smt.value_counts())

x_pred=rfc.predict(X_train_smt)
y_pred=rfc.predict(X_test)
print('After',counter)
from sklearn.metrics import accuracy_score
print("\nModel Training accuracy with 10 decision trees : ",format(accuracy_score(y_train_smt,x_pred)) )
print("\nModel Testing accuracy with 10 decision trees : ",format(accuracy_score(y_test,y_pred)) )
```

# By using n_estimaters

```python
clf = RandomForestClassifier(n_estimators=100, random_state=50)
clf.fit(X_train, y_train)
sm =SMOTE()
X_train_sm, y_train_sm = sm.fit_resample(X_train, y_train)
y_pred_100 = clf.predict(X_test)
x_pred_100 = clf.predict(X_train_sm)

print(classification_report(y_test,y_pred_100))
print("------------------------\n")
print("Confusion Matrix : \n")
print(confusion_matrix(y_test,y_pred_100))

print("\nTraining Acuracy with 100 Estimators : ",format(accuracy_score(y_train_sm,x_pred_100)))
print(" Testing Accuracy with 100 Estimators : ",format(accuracy_score(y_test,y_pred_100)))
```

# **Conclusion :**

1. Borrowers with low FICO scores and a high number of inquiries in the last six months are less likely to meet the credit policy approval criteria
2. Borrowers with high debt-to-income (DTI) ratios and high interest rates are more likely to default on their loans, failing to fully repay within the specified time frame.
3. We built several models to predict the likelihood of loan default for assessing borrowers' credit risk.
4. The logistic regression model achieved the best accuracy, followed closely by the SVM model, outperforming other models.
5. Based on this performance, we decided that logistic regression is the best fit for our dataset.
